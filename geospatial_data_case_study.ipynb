The below code explains the comparision of Hospitals and Pharmacy in particular city

```python
import geopandas as gpd
import matplotlib.pyplot as plt

# Load GeoJSON files into GeoDataFrames
hospitals = gpd.read_file("/content/export (1) (1).geojson")
pharmacy = gpd.read_file("/content/export (2).geojson")

# Count the number of features (rows) in each GeoDataFrame
hospital_count = len(hospitals)
pharmacy_count = len(pharmacy)

# Create a bar chart
categories = ['Hospitals', 'pharmacy']
counts = [hospital_count,pharmacy_count]

plt.bar(categories, counts, color=['red', 'blue'])
plt.xlabel('Category')
plt.ylabel('Count')
plt.title('Distribution of Hospitals and pharmacy in paris')
plt.show()
```

From the analysis ,Hospitals are very less compared to Pharmacy

```python
import matplotlib.pyplot as plt

# Filter the GeoDataFrame for hospitals and grocery stores
hospitals = hospital[hospital['amenity'] == 'hospital']
pharmacy = pharmacy[pharmacy['amenity'] == 'pharmacy']

# Count the number of each amenity type
hospital_counts = hospitals.shape[0]
pharmacy_counts = pharmacy.shape[0]

# Plot a pie chart
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Pie chart
axes[0].pie([hospital_counts, pharmacy_counts], labels=['Hospitals', 'pharmacy'], autopct='%1.1f%%', startangle=90)
axes[0].axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
axes[0].set_title('Distribution of Hospitals and pharmacy')

# Bar chart
amenity_types = ['Hospitals', 'pharmacy']
counts = [hospital_counts,pharmacy_counts]

axes[1].bar(amenity_types, counts, color=['red', 'blue'])
axes[1].set_ylabel('Count')
axes[1].set_title('Comparison of Hospitals and pharmacy')

plt.show()
```


Two Pandas DataFrames, hospital and pharmacy, are created using sample data. These DataFrames represent different amenities (e.g., hospitals, pharmacies).
Count Amenities:

The number of occurrences of each amenity type is calculated using the value_counts method. Specifically, the script counts the number of hospitals and pharmacies.
Plot Pie Charts:

Two pie charts are plotted side by side using Matplotlib. The first pie chart illustrates the overall distribution of amenities, showing the percentage of each amenity type in the entire dataset. The second pie chart focuses on hospitals and pharmacies, depicting the percentage distribution of these two specific amenities.
Pie Chart Customization:

The autopct parameter is used to display the percentage on each slice of the pie chart. The startangle parameter sets the initial rotation angle of the pie chart. The aspect ratio is set to be equal to ensure that the pie charts are drawn as circles.
Display Charts:

Finally, the plt.show() command is used to display the two pie charts in a single figure.

```python
import matplotlib.pyplot as plt
import pandas as pd

# Sample data (replace this with your actual data)
hospital_data = {'amenity': ['hospital', 'hospital', 'pharmacy']}
pharmacy_data = {'amenity': ['pharmacy', 'pharmacy', 'library']}

hospital = pd.DataFrame(hospital_data)
pharmacy = pd.DataFrame(pharmacy_data)

# Count the number of hospitals and schools
hospital_count = len(hospital[hospital['amenity'] == 'hospital'])
pharmacy_count = len(pharmacy[pharmacy['amenity'] == 'pharmacy'])

# Plot a pie chart
fig, ax = plt.subplots(1, 2, figsize=(12, 6))

# Pie chart for overall amenity distribution
ax[0].pie(hospital['amenity'].value_counts(), labels=hospital['amenity'].value_counts().index, autopct='%1.1f%%', startangle=90)
ax[0].axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
ax[0].set_title('Distribution of Amenities')

# Pie chart for hospitals and schools
ax[1].pie([hospital_count,pharmacy_count], labels=['Hospitals', 'pharmacy'], autopct='%1.1f%%', startangle=90)
ax[1].axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
ax[1].set_title('Distribution of Hospitals and pharmacy')

plt.show()
```
